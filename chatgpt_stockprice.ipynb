## Prompt for ChatGPT:
give me the python code to plot the following data:
https://raw.githubusercontent.com/tewei0328/teach-database/main/TXF1_1mk_2024_0322_0329.csv

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
url = "https://raw.githubusercontent.com/tewei0328/teach-database/main/TXF1_1mk_2024_0322_0329.csv"
data = pd.read_csv(url)

# Convert the 'Time' column to datetime format
data['Time'] = pd.to_datetime(data['date'])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['close'], label='Close Price', color='blue')
plt.title('TXF1 Close Price (2024-03-22 to 2024-03-29)')
plt.xlabel('Time')
plt.ylabel('Close Price')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
