## chatgpt prompt: 請給我可讀取以下連結資料, 並視覺化的python程式碼:
https://raw.githubusercontent.com/tewei0328/teach-database/main/banking33.csv

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取CSV文件
url = 'https://raw.githubusercontent.com/tewei0328/teach-database/main/banking33.csv'
data = pd.read_csv(url)

# 檢查數據的前幾行
print(data.head())

# 簡單的視覺化示例：繪製年齡的直方圖
plt.figure(figsize=(10, 6))
plt.hist(data['年'], bins=30, edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 簡單的視覺化示例：繪製收入與貸款金額的散點圖
plt.figure(figsize=(10, 6))
plt.scatter(data['資產總額-新臺幣百萬元'], data['存款餘額-新臺幣百萬元'], alpha=0.5)
plt.title('Income vs Loan Amount')
plt.xlabel('Income')
plt.ylabel('Loan Amount')
plt.grid(True)
plt.show()
```
